```python
Gender #let us first consider this feature and explore our data 
```

```python
#step 1 
df.columns 
```

```python
#Step 2 
ax= sns.countplot(x='Gender',data=df) 
#sns is seaborne and data is our data frame  
#we addd the below to get values to our bars in graphs/charts 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Step 3 
sales_gen= df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False )  

#GroupBy in Python and SQL is used to separate identical data into groups to allow for further aggregation and analysis. 
#A GroupBy in Python is performed using the pandas library groupby() function .

sns.barplot(x= 'Gender', y='Amount', data = sales_gen )
# we plot a graph taking Gender and Amount in X And Y axis resplectively
```

```python
Age 
# the second criteria 
```

```python
df.columns # we use the step 1 
```

```python
# same as Step 2 
ax=sns.countplot (x='Age Group', data=df, hue= 'Gender')  # we execute the Step 2 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Step 3 
sales_age= df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False )  

sns.barplot(x='Age Group', y='Amount', data = sales_age )
```

```python
State 
# Criteria no 3 to check which state makes the most orders 
```

```python
df.columns #s1
```

```python
# s2 we are finding taotal order in staes 

sales_State= df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False ).head(7)
#you can change head values in order to get desired number of entries 
# now we set the size of our bar in the plot/graph
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State', y='Orders', data = sales_age )
```

```python
# s3 we now find total amount of sales in states
sales_State= df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False ).head(7)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State', y='Amount', data=sales_State)
```
